## Pandas export
1. to_csv
2. to_excel
3. to_html
4. to_json

```python
import pandas as pd
df = pd.DataFrame()
```

## 1. to_csv

```python
df = pd.read_csv('deliveries.csv')
df.head()
```

```python
temp_df = df.groupby(['batsman','batting_team'])['batsman_runs'].sum().reset_index()
```

```python
temp_df.to_csv('batsman.csv',index=False)
```

```python
pd.read_csv('batsman.csv').head()
```

## 2. to_excel

```python
df.head()
```

```python
new_df = df.groupby('batsman')['batsman_runs'].sum().reset_index()
```

```python
new_df.to_excel('batsman_runs.xlsx',index=False)
```

```python
pd.read_excel('batsman_runs.xlsx')
```

```python
new_df.to_excel('output.xlsx',sheet_name='batsman')
```

```python
new_df1 = df.pivot_table(index='batsman',columns='bowling_team',values='batsman_runs',aggfunc='sum')
```

```python
with pd.ExcelWriter('batbowl.xlsx') as writer:
  new_df.to_excel(writer,sheet_name='batsman')
  new_df1.to_excel(writer,sheet_name='batsman_runs')
```

## 3. to_html

```python
temp = df.query('batsman_runs==6').pivot_table(index='over',columns='ball',values='batsman_runs',aggfunc='sum')
```

```python
temp.to_html('sixes_heatmap.html')
```

```python
temp.head()
```

## 4. to_json

```python
new = df.groupby(['batting_team','batsman'])['batsman_runs'].sum().unstack()
```

```python
new.head()
```

```python
new.to_json('ipl.json')
```
